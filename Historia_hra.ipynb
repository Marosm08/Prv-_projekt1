# Hra - História

Hra, pri ktorej je cieľom uhádnuť či tvrdenie je pravda alebo lož.

**Postup**

1. Vyber si náhodné číslo od 1-10
2. Pod číslami sa vždy nachádza téma a tvrdenie na danú tému.
3. Vašim cieľom je rozhodnúť či to tvrdenie je pravda alebo lož.
4. Za správne určenie pravdivosti tvrdenia sa dostanú 3 body, za nesprávne určenie sa uberie 1 bod.
5. Hráč, ktorý má najviac bodov na konci hry vyhráva.




Počet hráčov - 2

Dĺžka hry - cca. 10 minút





```python
# prompt: Naprogramuj mi hru - pravda alebo lož, bude mať 10 náhodných čísel, pod ktorými sa ukrývajú témy a otázky o histórii, cieľom hry je rozhodnúť o pravdivosti tvrdení - pravda/lož, za správne určenie pravdivost sa dostanú 3 body, za nesprávne určenie pravdivosti tvrdeni sa odoberie 1 bod. Hráč, ktorý má najviac bodov vyhráva.

import random

questions = {
    1: {
        "theme": "Staroveký Egypt",
        "question": "Pyramídy v Gíze boli postavené ako hrobky pre faraónov.",
        "answer": True
    },
    2: {
        "theme": "Rímska ríša",
        "question": "Julius Caesar bol prvý rímsky cisár.",
        "answer": False
    },
    3: {
        "theme": "Stredovek",
        "question": "Veľká morová epidémia v 14. storočí zabila približne tretinu populácie Európy.",
        "answer": True
    },
    4: {
        "theme": "Renesancia",
        "question": "Leonardo da Vinci namaľoval Mona Lisu.",
        "answer": True
    },
    5: {
        "theme": "Veľké geografické objavy",
        "question": "Krištof Kolumbus objavil Ameriku v roku 1492.",
        "answer": True
    },
    6: {
        "theme": "Francúzska revolúcia",
        "question": "Bastila bola dobytá 14. júla 1789.",
        "answer": True
    },
    7: {
        "theme": "Prvá svetová vojna",
        "question": "Začiatkom Prvej svetovej vojny bol atentát na arcivojvodu Františka Ferdinanda.",
        "answer": True
    },
    8: {
        "theme": "Druhá svetová vojna",
        "question": "Hitler spáchal samovraždu v bunkri v Mníchove.",
        "answer": False
    },
    9: {
        "theme": "Studená vojna",
        "question": "Berlínsky múr padol v roku 1991.",
        "answer": False
    },
    10: {
        "theme": "Slovenské dejiny",
        "question": "Prvá Slovenská republika vznikla v roku 1940.",
        "answer": False
    }
}

def play_game():
    score = 0
    used_numbers = []

    print("Vitajte v hre Pravda alebo Lož - História!")
    print("Cieľom je určiť, či je tvrdenie pravdivé.")
    print("Za správnu odpoveď získate 3 body, za nesprávnu sa vám 1 bod odoberie.")

    for _ in range(10):
        while True:
            question_number = random.randint(1, 10)
            if question_number not in used_numbers:
                used_numbers.append(question_number)
                break

        question_data = questions[question_number]
        print(f"\nTéma: {question_data['theme']}")
        print(f"Tvrdenie: {question_data['question']}")

        while True:
            user_answer = input("Je to pravda alebo lož? (pravda/loz): ").lower()
            if user_answer in ["pravda", "loz"]:
                break
            else:
                print("Neplatná odpoveď. Prosím, zadajte 'pravda' alebo 'loz'.")

        is_true = question_data["answer"]

        if (user_answer == "pravda" and is_true) or (user_answer == "loz" and not is_true):
            print("Správne!")
            score += 3
        else:
            print("Nesprávne.")
            score -= 1

        print(f"Aktuálne skóre: {score}")

    print("\nHra sa skončila!")
    print(f"Vaše konečné skóre je: {score}")

    if score > 0:
        print("Gratulujeme, vyhrali ste!")
    elif score < 0:
        print("Nabudúce to určite pôjde lepšie.")
    else:
        print("Hra skončila nerozhodne.")

# prompt: Naprogramuj mi hru - pravda alebo lož, bude mať 10 náhodných čísel, pod ktorými sa ukrývajú témy a otázky o histórii, cieľom hry je rozhodnúť o pravdivosti tvrdení - pravda/lož, za správne určenie pravdivost sa dostanú 3 body, za nesprávne určenie pravdivosti tvrdeni sa odoberie 1 bod. Hráč, ktorý má najviac bodov vyhráva.

import random

questions = {
    1: {
        "theme": "Staroveký Egypt",
        "question": "Pyramídy v Gíze boli postavené ako hrobky pre faraónov.",
        "answer": True
    },
    2: {
        "theme": "Rímska ríša",
        "question": "Julius Caesar bol prvý rímsky cisár.",
        "answer": False
    },
    3: {
        "theme": "Stredovek",
        "question": "Veľká morová epidémia v 14. storočí zabila približne tretinu populácie Európy.",
        "answer": True
    },
    4: {
        "theme": "Renesancia",
        "question": "Leonardo da Vinci namaľoval Mona Lisu.",
        "answer": True
    },
    5: {
        "theme": "Veľké geografické objavy",
        "question": "Krištof Kolumbus objavil Ameriku v roku 1492.",
        "answer": True
    },
    6: {
        "theme": "Francúzska revolúcia",
        "question": "Bastila bola dobytá 14. júla 1789.",
        "answer": True
    },
    7: {
        "theme": "Prvá svetová vojna",
        "question": "Začiatkom Prvej svetovej vojny bol atentát na arcivojvodu Františka Ferdinanda.",
        "answer": True
    },
    8: {
        "theme": "Druhá svetová vojna",
        "question": "Hitler spáchal samovraždu v bunkri v Mníchove.",
        "answer": False
    },
    9: {
        "theme": "Studená vojna",
        "question": "Berlínsky múr padol v roku 1991.",
        "answer": False
    },
    10: {
        "theme": "Slovenské dejiny",
        "question": "Prvá Slovenská republika vznikla v roku 1940.",
        "answer": False
    }
}

def play_game():
    score = 0
    used_numbers = []

    print("Vitajte v hre Pravda alebo Lož - História!")
    print("Cieľom je určiť, či je tvrdenie pravdivé.")
    print("Za správnu odpoveď získate 3 body, za nesprávnu sa vám 1 bod odoberie.")

    for _ in range(10):
        while True:
            question_number = random.randint(1, 10)
            if question_number not in used_numbers:
                used_numbers.append(question_number)
                break

        question_data = questions[question_number]
        print(f"\nTéma: {question_data['theme']}")
        print(f"Tvrdenie: {question_data['question']}")

        while True:
            user_answer = input("Je to pravda alebo lož? (pravda/loz): ").lower()
            if user_answer in ["pravda", "loz"]:
                break
            else:
                print("Neplatná odpoveď. Prosím, zadajte 'pravda' alebo 'loz'.")

        is_true = question_data["answer"]

        if (user_answer == "pravda" and is_true) or (user_answer == "loz" and not is_true):
            print("Správne!")
            score += 3
        else:
            print("Nesprávne.")
            score -= 1

        print(f"Aktuálne skóre: {score}")

    print("\nHra sa skončila!")
    print(f"Vaše konečné skóre je: {score}")

    if score > 0:
        print("Gratulujeme, vyhrali ste!")
    elif score < 0:
        print("Nabudúce to určite pôjde lepšie.")
    else:
        print("Hra skončila nerozhodne.")

if __name__ == "__main__":
    play_game()
```
